##  INTRODUCTION

The aim of this project was to generate an interactive program which allows users to perform their own data science analysis on a dataset of COVID-19 outcomes and demographic factors for UK local authorities. This project aims to enable local authorities to explore and visualize the link between demographic characteristics of local authorities and COVID-19 outcomes, and predict cases based upon this data. 

In order to create our program, we gathered relevant, most up to date demographic and Covid-19 related data as of March 17th data for 302 regions of the UK from the following sources: Office of National Statistics: population estimates(1) and population profiles(2), and Covid-19 data from .gov.uk(3).

The demographic data included datapoints(columns) such as deprivation, employment levels, ethnicity, age, population density, etc. The relevant Covid-19 information we collected included: cumulative death, population vaccinated 1st, 2nd, and 3rd doses, demographic data etc. The separate CSV files are contained in our Github repository.  Our program was designed to understand how Covid-19 data for UK regions relates to the demographic information for those regions. To achieve this, we generated a code base to help us achieve the following core goals:

•	Data Modelling (Regression_models.py, etc.)

## VARIABLE FINDER

Variable finder is a python file which contains a class which confirms that a variable that the user inserted exists in the dataset. For example, in Info.py the user is prompted to enter a variable name that they want to see the head, tail or mean for; and then variable finder confirms if the variable exists in the dataset. If the variable does not exist, a while True loop repeats and the user is prompted to enter a variable name again. The class also takes an input of the variable type that the user must input. For example, if the variable is the X variable of a graph, the class will ask the user “Which X variable do you want to look at?”. It was chosen to make variable finder a class because the user is regularly asked to input a variable name, and using a class reduces the amount of repeated code, and allows its use across different files.

## DATA MODELLING

Our program also performs AI and Machine learning data training and modelling for the user. This is a useful tool for prediction, making statistical inferences based on the data, and understanding which model fits best to data. We believe decision makers and policy makers will find this useful. 

In the Regression_models.py file, we defined 6 sub-functions within a function called regression (). 3 of the subfunctions run regression algorithms on the data, and the regression algorithms are linear regression, polynomial regression, and decision tree, respectively. The other sub-functions are the CompareAllModels(), SelectTargetVariable(), and traintestsplit(). These help the user compare performance of each model and select a target variable as well as split data before running the regression algorithms through the data. 

In order to run the regression algorithms, all the user has to do is specify a given y variable, which is then dropped from our data in order to form the separate y variable. The columns that are left in the table form the x variable. This specific variable conceptualization operation is done through variable_finder.py which then feeds Regression_models.py file, that runs our 3 regression algorithms (written within functions).  Within Regression_models.py, before running a function, the user is asked to choose which of the 3 regression algorithms they want to fit to the data or if they want to run all 3 and compare. Specific code below: 

 **regression_type = input("Do you want to use a linear regression model (L), Polynomial regression (P), Decision Tree Regression (D), compare all (C) or return to covid_project.py (main)?**

Note that each input option given above is a function on its own within Regression_models.py, except covid_project.py, which is another code file itself (root code file).  

Once an applicable input is given by the user, Regression_models.py runs by first splitting the X and Y data that has been fed into it from variable_finder.py into training and test data by running train_test_split function and entering arguments such as test_size and random_state for specifying which fraction of data becomes test data and for reproducibility, respectively. The split data (Xtr, Xtest, Ytr, Ytest) become the arguments of the functions that will help us run the different algorithms and compare them. 

In order to prevent our program from crashing, we implemented a while true loop with if, else statements that will only run the functions within Regression_models.py if the user enter the correct input value. See an excerpt from regression_models.py below: 

```python
while True:
        regression_type = input("Do you want to use a linear regression model (L), Polynomial regression (P), Decision Tree Regression (D), compare all (C) or return to covid_project.py (main)? ")
        if regression_type == "L":
            linear(Xtraining, Xtesting, Ytraining, Ytesting, Parameters, Target)
            info_stay = input("Do you want to stay in regression models? (Y or N) ")
            if info_stay == "Y":
                continue
            elif info_stay == "N":
                break
            else:
                print("Invalid input")
                continue
```

What is happening in the box above is that we set conditions for the value of the input parameter a user enters to run the linear regression function. If user enters L in response to the regression_type input variable, the program is instructed to run a linear regression model denoted with the linear() function. Once the linear regression algorithm runs, the user is then asked if they wish to continue in regression models and run other models. If the answer is no, the program breaks. If yes, the program continues. If user enters something else, the program prints ‘Invalid input’. 

We also apply a While True loop to the train_test_split() function for the purpose of ensuring our test size input values are between 0.1 and 0.5, and to make sure the input value entered to represent random state is actually an integer. The program minimizes the possibility of crashing by only running when the right input has been entered.
